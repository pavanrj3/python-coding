```python
#create a python code for waiter to serve food and also we know what dish we ordered
from abc import ABC, abstractmethod

class Dish(ABC):
    @abstractmethod
    def serve(self):
        pass
class Pasta(Dish):
    def serve(self):
        print("Serving Pasta.")
class Burger(Dish):
    def serve(self):
        print("Serving Burger.")
class Salad(Dish):
    def serve(self):
        print("Serving Salad.")
def dishSelection(dish_type):
    if dish_type == "pasta":
        return Pasta()
    elif dish_type == "burger":
        return Burger()
    else:
        return Salad()
waiter1 = dishSelection("pasta")
waiter1.serve()
waiter2 = dishSelection("burger")
waiter2.serve()
waiter3 = dishSelection("salad")
waiter3.serve()
```

```python
#linkedlist in python (add and travesing)
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None
class LinkedList:
    def __init__(self):
        self.head=None
    def print_LL(self):
        if self.head is None:
            print("Linked List is empty: ")
        else:
            n=self.head
            while n is not None:
                print(n.data)
                n=n.ref

    def add_begin(self,data):
        new_node=Node(data)
        new_node.ref=self.head
        self.head=new_node

    def add_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            n=self.head
            while n.ref is not None:
                n=n.ref
            n.ref=new_node

LL1=LinkedList()
LL1.add_begin(10)
LL1.add_end(30)
LL1.add_begin(20)
LL1.add_end(1)
LL1.print_LL()
```
